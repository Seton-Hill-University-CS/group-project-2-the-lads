```python
# importing necessary libraries
import random
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


# %%
#Create two lists of data (X and Y) - randomize 100 values in each list from a range of 1 to 250
X = [random.randint(1,250) for i in range (1,100)]
Y = [random.randint(1,250) for i in range (1,100)]
data = np.column_stack((X, Y))

# %%
#Plot the data points on a graph

# plot points
plt.scatter(X,Y, marker = 'x', color = 'green')

# name x label
plt.xlabel('x_axis')

# name y label
plt.ylabel('y_axis')

# create title
plt.title('Relationship between X and Y')

#show graph
plt.show()

# %%
#Using a distance formula, determine if more than one cluster (or grouping) appears within the data.
k_values = range (1,99)
#sum of squared distances within clusters

ssd = []

for k in k_values:
    model = KMeans(n_clusters = k)
    model.fit(data)
    ssd.append(model.inertia_)

# %%
#plot the elbow method to find an optimal number of clusters

#developing figure size
plt.figure(figsize=(8, 6))

#ploting the x and y values in K_means
plt.plot(k_values, ssd, marker='o')

#title of the graph
plt.title('Elbow Method For Optimal K')

#x-axis label
plt.xlabel('Number_of_Clusters')

#y-axis label
plt.ylabel('Sum_of_squared_distances')

#showing the graph
plt.show()

# %%
 # Set the optimal K based on the elbow method
# We chose an optimal value of K = 10 due to the leveling off of the graph at that point
optimal_k = 10

kmeans = KMeans(n_clusters=optimal_k)
kmeans.fit(data)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Visualize the clusters
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], s=200, marker='X', c='red', label='Centroids')
plt.title(f'Clustering with K={optimal_k}')
plt.legend()
plt.show()
```
